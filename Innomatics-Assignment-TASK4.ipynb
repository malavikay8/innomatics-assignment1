## Problem 1: Mobile Recharge Validation System


```python
# Valid plans
valid_plans = [199, 299, 399, 599]
def validate_recharge(amount):
  if amount < 50:
    print("Amount must be atleast ₹50")
  if amount not in valid_plans:
    print("Invalid plan selected")
  else:
    return True
while True:
  amount = int(input("Enter recharge amount: "))
  if validate_recharge(amount):
    print("Recharge Successul")
    break
  else:
    print("Please try again")
```

## Problem 2: Inventory Reorder Alert System


```python
inventory = {
    "Mobile": 20,
    "Laptop": 10,
    "Headphones": 5,
    "Charger": 25
}
def check_stock(product, quantity):
  if quantity < 15:
    print(f"{product} : Reorder Alert")
  else:
    print(f"{product} : Stock OK")

for product, quantity in inventory.items():
  check_stock(product, quantity)
```

## Problem 3:Student Result Processing System


```python
def process_result(marks):
    total = 0
    # calculate total
    for m in marks:
        total += m

    # calculate average
    average = total / len(marks)

    if average >= 50:
        print("Pass")
    else:
        print("Fail")

student_marks = [60, 55, 70, 45]

result = process_result(student_marks)

print("Average Result:", result)
```

## Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak_hour):
    base_fare = 50
    rate_per_km = 12

    fare = base_fare + (rate_per_km * distance)

    # add 25% extra during peak hour
    if peak_hour == "yes":
        fare = fare * 1.25
    return fare

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)

    print(f"Estimated Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ").lower()

    if retry != "yes":
        print("Thank you for using Cab Fare Estimator!")
        break
```

## Problem 5: Employee Attendance Eligibility Checker


```python
def check_eligibility(attendance):
  present_count = 0
  for status in attendance:
    if status == "P":
      present_count += 1
  percentage = present_count / len(attendance) * 100
  if percentage >= 75:
    return "Eligible"
  else:
    return "Not Eligible"

attendance_list = ["P", "P", "A", "P", "P"]
result = check_eligibility(attendance_list)
print("Attendance Status:", result)
```
